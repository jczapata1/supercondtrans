**Fields**

```python
from programs.tdgl_fields import TDGLFields
```

```python
profiles = ['Uniform', 'Plateau', 'Domains']
gauge    = 'Landau-y'
```

```python
for profile in profiles:

    # Print
    print(f'Profile: {profile}')
    
    # Fields
    fld = TDGLFields(folder=f'./input/Fields/{profile}')
    
    # Load
    fld.read_parameters(file=f'./input/Fields/{profile}/Input.in') 
    fld.load_device(file='./input/Device/Device.h5')   

    # Make
    fld.make_magnetic_field(profile=profile)  
    fld.make_vector_potential(gauge=gauge) 

    # Plot
    fld.plot_magnetic_field()   
    fld.plot_vector_potential() 
```

**Quick Guide**

`profile:`

* Uniform: uniform magnetic field profile.
* Plateau: plateau magnetic field profile.
* Domains: periodic stripe domain magnetic field profile.

`gauge:`

* Landau-y: when $B_{z} = B_{z}(y)$ (Uniform, Plateau, and Domains $\theta=0$).
* Landau-x: when $B_{z} = B_{z}(x)$ (Domains $\theta=\pi/2$).
* Rotated: when $B_{z} = B_{z}(x,y)$ ($\theta$ Arbitrary).
