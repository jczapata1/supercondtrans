**Animation**

```python
from programs.tdgl_animation import TDGLAnimation
```

```python
simulation = 'Static'
profile    = 'Uniform'
anim       = TDGLAnimation(folder=f'./output/{simulation}/{profile}')
```

```python
# Load
anim.load_device(file='./input/Device/Device.h5')
anim.load_solution(file=f'./output/{simulation}/{profile}/Solution.h5')

# Animation
anim.snapshots(n=20)
anim.animation()
anim.compress()
```

**Quick Guide**

`simulation:`

* Static: TDGL solve without transport current (point mode).
* Dynamic: TDGL solve with applied transport current (point mode).
* Sweep: TDGL solve with applied transport current (sweep mode).

`profile:`

* Uniform: uniform magnetic field profile.
* Plateau: plateau magnetic field profile.
* Domains: periodic stripe domain magnetic field profile.
