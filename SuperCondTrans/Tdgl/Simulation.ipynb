**Simulation**

```python
from programs.tdgl_simulation import TDGLSimulation
from programs.tdgl_animation import TDGLAnimation
```

```python
# Simulation
simulation = 'Static'
folder     = f'./output/{simulation}'
sim        = TDGLSimulation(folder=folder)
```

```python
# Load
sim.load_setup(folder='./input/Setup')
sim.set_current(file=f'./input/Current/{simulation}.in')
```

```python
# Run
if simulation in ('Static', 'Dynamic'):

    # Simulation
    sim.point(simulation=simulation, seed_solution=None)
    sim.plot_psi()
    sim.plot_phase()
    sim.plot_vorticity()
    sim.plot_normal_current()
    sim.plot_super_current()
    sim.plot_scalar_potential()

    # Animation
    anim = TDGLAnimation(folder=folder)
    anim.device = sim.device
    anim.solution = sim.solution
    anim.snapshots()
    anim.animation()

elif (simulation == 'Sweep'):
    
    # Simulation
    sim.sweep(simulation=simulation, seed=False)
    sim.plot_current_voltage()

    # Animation
    for (i, solution) in enumerate(sim.solutions):
        
        anim_i = TDGLAnimation(folder=f'{folder}/Data/{i}')
        anim_i.device = sim.device
        anim_i.solution = solution
        anim_i.snapshots()
        anim_i.animation()

    sim.compress()

else:
    raise ValueError(f"Invalid Simulation Mode!. Use 'Static', 'Dynamic', or 'Sweep'.")
```

**Quick Guide**

`simulation:`

* Static: TDGL solve without transport current (point mode).
* Dynamic: TDGL solve with applied transport current (point mode).
* Sweep: TDGL solve with applied transport current (sweep mode).
