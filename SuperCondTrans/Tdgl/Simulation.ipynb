**Simulation**

```python
from programs.tdgl_simulation import TDGLSimulation
```

```python
simulation = 'Static'
profile    = 'Uniform'
sim        = TDGLSimulation(folder=f'./output/{simulation}/{profile}')
```

```python
# Load
sim.read_parameters(file='./input/Input.in')
sim.load_device(file='./input/Device/Device.h5')
sim.load_magnetic_field(file=f'./input/Fields/{profile}/Magnetic_Field.h5')
sim.load_vector_potential(file=f'./input/Fields/{profile}/Vector_Potential.h5')
sim.set_current(file=f'./input/Current/{simulation}/Input.in')

# Run
if simulation in ('Static', 'Dynamic'):
    sim.point(simulation=simulation)
    sim.plot_psi()
    sim.plot_phase()
    sim.plot_vorticity()
    sim.plot_scalar_potential()
    
elif (simulation == 'Sweep'):
    sim.sweep(simulation=simulation, seed=False)
    sim.compress()
    sim.plot_current_voltage()
    
else:
    raise ValueError(f"Invalid Simulation Mode!. Use 'Static', 'Dynamic', or 'Sweep'.")

# Voltage
sim.voltage
```

**Quick Guide**

`simulation:`

* Static: TDGL solve without transport current (point mode).
* Dynamic: TDGL solve with applied transport current (point mode).
* Sweep: TDGL solve with applied transport current (sweep mode).

`profile:`

* Uniform: uniform magnetic field profile.
* Plateau: plateau magnetic field profile.
* Domains: periodic stripe domain magnetic field profile.
